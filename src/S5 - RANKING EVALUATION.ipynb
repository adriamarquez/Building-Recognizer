# Construct BoW vector + Save Bow Descriptors

##### En la funció construct_BoW_vector.py es tracta de que aquesta reotrni un vector amb el número de descriptors situats en cadascun dels vectors normalitzats entre els valors 0 i 1 per desprès guardar el 100 primers en un diccionari on en cada fila estiguessin associats al identificador de imatge corresponent. El següent codi conté el compute_assignments, el construct_bow_vector i la creació dels diccionaris tant de train com de val.

```python
#CREACIÓ DICCIONARI DE TRAIN
FE.save_bow(centroides,dsc_ind_train,"train",nclusters) #calculem i dessem els bow de les imatges d'entrenament
# CÀLCUL DELS DESCRIPTORS DE VALIDACIÓ + CREACIÓ DICCIONARI DE VALIDACIÓ
_, dsc_ind_val = FE.get_descriptors("../TerrassaBuildings900/val/images") #obtenim els descriptors de validació
FE.save_bow(centroides,dsc_ind_val,"val",nclusters) #guardem els bow de les imatges de validació
```

##### L'output de la funció seran dos diccionaris retornats amb la id de la imatge i el 100 primers descriptors normalitzats entre 0 i 1. S'ens crearan dos diccionaris: bow_train.p i bow_val .p

# Ranking
