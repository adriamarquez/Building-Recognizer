### Gràfiques Performance/num. nodes i Temps/num. nodes

Hem realitzat una evaluació del ranking per a diferents mètodes d'extracció i diferents numeros de clusters. Tenint com a resultats els següents que mostrem a les gràfiques. Ens ha sorprés que rootSIFT sigui el que pitjor retrieval dona.

```python
import numpy as np
import matplotlib.pyplot as plt

performance = [300,500,800,1000,1200,1400,1600]
rootsift = [0.1257,0.1503,0.1675,0.1815,0.1819,0.1398,0.1340]
sift=[0.2245,0.2287,0.2353,0.2423,0.2380,0.1917,0.2534]
surf=[0.2534,0.2404,0.2372,0.2833,0.2856,0.28920,0.2893]
rootsifttime=[63.73,119.72,112.11,140.39,184.25,136.58,163.16]
sifttime=[96.05,145.18,209.37,242.10,230.06,142.49,344.17]
surftime=[105.38,92.65,110.12,231.99,232.80,273.68,304.45]

figure(figsize(10, 5))
plt.subplot(1, 2, 1)
plt.plot(performance, rootsift,label='RootSIFT')
plt.plot(performance, sift,label='SIFT')
plt.plot(performance, surf, label='SURF')
plt.xlabel('Num.Clusters')
plt.ylabel('MAP')
plt.title('MAP/Num.Clusters')
plt.legend(loc="upper left")

plt.subplot(1, 2, 2)
plt.plot(performance, rootsifttime,label='RootSIFT')
plt.plot(performance, sifttime,label='SIFT')
plt.plot(performance, surftime,label='SURF')
plt.xlabel('Num.Clusters')
plt.ylabel('Time')
plt.title('Time/Num.Clusters')

plt.legend(loc="upper left")
plt.show()
```
